```python
from bs4 import BeautifulSoup as bsoup
from urllib.request import urlopen
            
import pandas as pd
import numpy as np
```

```python
def make_soup(url):
    thepage=urlopen(url)
    soupdata=bsoup(thepage, "html.parser")
    return soupdata

soup= make_soup("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
albumdatasaved=""
# find all table ,get the first
table = soup.find_all('table', class_="wikitable")[0]  # Only use the first table
# iter over it
for record in table.findAll('tr'):
    albumdata=""
    for data in record.findAll('td'):
        albumdata=albumdata+","+data.text
    albumdatasaved=albumdatasaved+"\n"+albumdata[1:]

rows = [x for x in albumdatasaved.split('\n') if x!='']

alldata = []
for x in rows:
    temp = x.split(',')
    if temp[2]=='Not assigned' and temp[1]!='Not assigned':
        temp[2]=temp[1]
        alldata.append(temp)
    elif temp[1]!='Not assigned':
        alldata.append(temp)
        
for idx1,i in enumerate(alldata):
    for idx2,j in enumerate(alldata):
        if j!=i and idx2>idx1 and j[0]==i[0]:
            alldata[idx1][2]+=(','+alldata[idx2][2])
            del alldata[idx2]
```

```python
df = pd.DataFrame(data=alldata)
columns = ['Postal Code', 'Borough','Neighborhood']
df.columns = columns
df
```
